```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.mixture import GMM
from matplotlib.colors import LogNorm
```

Load feature names.

```python
feature_names = []
with open('mushroom.map') as fmap:
    for line in fmap:
        [index, name, junk] = line.split()
        feature_names.append(name)

print 'Loaded feature names:', len(feature_names)
```

Load data. The data is sparse in the input file, but there aren't too many features, so we'll use a dense representation, which is supported by all sklearn objects.

```python
X, Y = [], []

with open('mushroom.data') as fdata:
    for line in fdata:
        items = line.split()
        Y.append(int(items.pop(0)))
        x = np.zeros(len(feature_names))
        for item in items:
            feature = int(item.split(':')[0])
            x[feature] = 1
        X.append(x)

# Convert these lists to numpy arrays.
X = np.array(X)
Y = np.array(Y)

# Split into train and test data.
train_data, train_labels = X[:7000], Y[:7000]
test_data, test_labels = X[7000:], Y[7000:]

# Check that the shapes look right.
print train_data.shape, test_data.shape
```

(1) Run Principal Components Analysis on the data. Show what fraction of the total variance in the training data is explained by the first k principal components, for k in [1, 50].

```python
## functions accessible via global memory
def line(length=80):
    line = "-"
    return(line * length + '\n')

#small transcoder function
def trans_colors(x, dict):
    return dict[x]

  
    #define a small helper function to compare values in p5 & p6
def compare(x,ps,ns):
    if ps[x] > ns[x]:
        return 1
    else:
        return 0
```

```python
def P1(comp=50):
### STUDENT START ###
    pca = PCA(n_components= comp)
    pca.fit(train_data)
    
    total_var = np.sum(pca.explained_variance_ratio_[0:comp])
    
    print "PCA for %i" %comp + " components explains %f" %total_var + " total variance"
    print line()
    
    var = 0
    for i in range(0,comp):
        #iterate through the explained_variance_ratio_ array and accumulate values
        var += pca.explained_variance_ratio_[i]
        print "Variance explained by first %i principal components is %f" %(i+1, var)
    
### STUDENT END ###
P1()
```

(2) PCA can be very useful for visualizing data. Project the training data down to 2 dimensions and plot it. Show the positive (poisonous) cases in blue and the negative (non-poisonous) in red. Here's a reference for plotting: http://matplotlib.org/users/pyplot_tutorial.html

```python
def P2(comp=2):
### STUDENT START ###
    
    pca = PCA(n_components= comp)
    pca.fit(train_data)
    td = pca.transform(train_data)
    
    #poisonous (1)==blue, nonpoisonous (0)==red
    color_dict = {1: 'blue', 0: 'red'}
    label_colors = np.asarray([trans_colors(x,color_dict) for x in train_labels])
    
    #plot data, use label colors to encode
    plt.title('Poisonous (Blue) vs Non Poisonous (Red) Mushrooms plotted with %i PCA Dimensionality Reduction'%comp, y = 1.05)
    plt.scatter (td[:,0], td[:,1], c=label_colors)

### STUDENT END ###
P2()
```

(3) Run KMeans with [1,16] clusters over the 2d projected data. Mark each centroid cluster and plot a circle that goes through the most distant point assigned to each cluster.

```python
def P3(comp=2, k=17):
### STUDENT START ###
    
    #initialize
    pca = PCA(n_components= comp)
    
    #fit PCA on training data
    pca.fit(train_data)
    
    #transform training data
    td = pca.transform(train_data)
    
    plt.figure(figsize=(20,20))
   
    for k in range (1,k):
    #run Kmeans algo 
        km = KMeans(n_clusters=k, init='k-means++')

        #Fit on PCA transformed training data
        clusters = km.fit(td) 

        #determine which subplot location to fill
        plt.subplot(4, 4, k)
        
        #create a title
        plt.title('KMeans k = %i'%k, y = 1.02)

        
        #plot clusters and color by label index
        plt.scatter(td[:,0], td[:,1], c=clusters.labels_, cmap='Accent')

        #plot centroids
        plt.scatter(clusters.cluster_centers_[:,0], clusters.cluster_centers_[:,1], color='orange', s = 100)

        # Calculate the furtherst point from each centroid, then plot it
        for x in range(0, clusters.cluster_centers_.shape[0]):
            # Distance: linalg.norm defaults to a Frobenius norm with L2 (Euclidian distance)
            dist = np.linalg.norm(td[clusters.labels_ == x] - clusters.cluster_centers_[x], axis=1)
            r = dist.max()
            # Add circle to plot by getting the current figure and then getting the current axes to add to
            plt.gcf().gca().add_artist(plt.Circle(xy=(clusters.cluster_centers_[x,0], clusters.cluster_centers_[x,1]), 
                                    radius=r, color='orange', alpha=0.1, zorder = 3))
            
### STUDENT END ###
P3()
```

(4) Fit a Gaussian Mixture Model for the positive examples in your 2d projected data. Plot the estimated density contours as shown here: http://scikit-learn.org/stable/auto_examples/mixture/plot_gmm_pdf.html#example-mixture-plot-gmm-pdf-py. Vary the number of mixture components from 1-4 and the covariance matrix type ('spherical', 'diag', 'tied', 'full').

```python
def P4(comp=2):
### STUDENT START ##
    fig, axes = plt.subplots(nrows=4, ncols=4, sharex=False, sharey=False, figsize = (16, 16))

    #initialize
    pca = PCA(n_components= comp)
    
    #fit PCA on training data
    pca.fit(train_data)
    
    #transform training data
    td = pca.transform(train_data)
    
    #covariance matrix types
    cov_type = ['spherical','diag','tied','full']
    
    #Select only training data that is labeled as positive
    pos_labels = td[train_labels==1]
    
    #For 1-4 GMM components
    for i in range(1,5):
        #For each covariance matrix type
        for idx,ct in enumerate(cov_type):
            #Fit a GMM with i components and covariance matrix j for the positive examples
            g = GMM(n_components=i,covariance_type=ct)            
            g.fit(pos_labels)

            #Plot the training data on the appropiate spot in the plot space
            ax = axes[i-1,idx]
            ax.scatter (pos_labels[:,0], pos_labels[:,1])

            # display predicted scores by the model as a contour plot
            x = np.linspace(-5.0, 5.0)
            y = np.linspace(-5.0, 5.0)
            X, Y = np.meshgrid(x, y)
            XX = np.array([X.ravel(), Y.ravel()]).T
            Z = -g.score_samples(XX)[0]
            Z = Z.reshape(X.shape)
            CS = ax.contour(X, Y, Z, norm=LogNorm(vmin=1.0, vmax=1000.0), levels=np.logspace(0, 3, 10))

            #Set title
            ax.set_title (str(i) + " components. Matrix Type: " + cov_type[idx], y = 1.02)

### STUDENT END ###

P4()
```

(5) Fit two 4-component full covariance GMMs, one for the positive examples and one for the negative examples in your 2d projected data. Predict the test examples by choosing the label for which the model gives a larger probability (use GMM.score). What is the accuracy?

```python
def P5(comp=2, gmm_comp = 4):
### STUDENT START ##

    #initialize
    pca = PCA(n_components = comp)
    
    #fit PCA on training data
    pca.fit(train_data)
    
    #transform training data
    td = pca.transform(train_data)
    #transform test data
    ted = pca.transform(test_data)

    #for both positive and negative classes
    for i in [0,1]:
        #Chooses positive or negative labels
        data = td[train_labels==i] 
        g = GMM(n_components=gmm_comp, covariance_type="full") 
        #Fits the GMMM to the data
        g.fit(data) 
        if i==0: 
            #If negative label, store in ns
            ns = g.score(ted)
        else: 
            #if positive label, store in ps
            ps = g.score(ted)
  
    #List comprehension to store predictions based on comparing which label is larger
    # user helper function "compare" defined above
    preds = np.asarray([compare(x,ps,ns) for x in range(0,test_data.shape[0])])
    
    #Print Accuracy
    print "Accuracy: %f" %(metrics.accuracy_score(test_labels, preds))
### STUDENT END ###
P5()
```

(6) Ideally, we'd like a model that gives the best accuracy with the fewest parameters. Run a series of experiments to find the model that gives the best accuracy with no more than 50 parameters. For example, with 3 PCA components and 2-component diagonal covariance GMMs, you'd have:

( (3 mean vector + 3 covariance matrix) x 2 components ) x 2 classes = 24 parameters

You should vary the number of PCA components, the number of GMM components, and the covariance type.

```python
def P6(max_parameters=50):
### STUDENT START ###
    #covariance matrix types
    cov_type = ['spherical','diag','tied','full']
   
    #Empty lists
    total_params,pc_comp,gmm_comp,cov_list,acc = ([] for i in range(5))

    #For 1-12 PCA components
    for comp in range(1,12):
        #Project the training data to p dimensions using PCA
        pca = PCA(n_components = comp)
        pca.fit(train_data)
        
        #Transform the training and test data
        td = pca.transform(train_data)
        ted = pca.transform(test_data)

        #For 1-8 GMM components
        for gg in range(1,8):
            #For each covariance matrix type
            for c in range(0,4):
                g = GMM(n_components=gg,covariance_type=cov_type[c]) 

                #for both positive and negative classes
                for i in [0,1]:
                    
                    #Chooses positive or negative labels
                    data = td[train_labels==i] 
                    g = GMM(n_components=gg, covariance_type="full") 
                    #Fits the GMMM to the data
                    g.fit(data) 
                    if i==0: 
                        #If negative label, store in ns
                        ns = g.score(ted)
                    else: 
                        #if positive label, store in ps
                        ps = g.score(ted)

                #Store Predictions
                preds = np.asarray([compare(x,ps,ns) for x in range(0,test_data.shape[0])])

                #Store results of each PCA/GMM/Covariance combination in a list
                if (comp+comp)*gg*2 <= max_parameters:
                    total_params.append((comp+comp)*gg*2)
                    pc_comp.append(comp)
                    gmm_comp.append(gg)
                    cov_list.append(cov_type[c])
                    acc.append(metrics.accuracy_score(test_labels,preds))

    #Stack all of the lists
    output = np.column_stack((total_params,pc_comp,gmm_comp,cov_list,acc))
    
    #Sorts them lists in descending order by accuracy (i.e. the 5 column in the stacked list)
    output=output[output[:,4].argsort()][::-1]
    
    #Prints the details of the most accurate models
    
    print "Top 10 Most Accurate Models"
    for i in output[:10]:
        print "Total Params: %s, PCA: %s, GMM: %s, Covariance: %s, Accuracy: %s" %(i[0],i[1],i[2],i[3],i[4])
### STUDENT END ###
P6()
```
